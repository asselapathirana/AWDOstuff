```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
from adjustText import adjust_text

# Load the Excel file and specify that the second row contains the column names
df = pd.read_excel('../../D1_jan2025/draft_january2025/Comparisons_2025_2020.xlsx', header=1)

# Extract relevant columns based on your data structure
df_2025 = df[['Country', 'Total KD5 Score of 20', 'KD5 Score']].copy()
df_2020 = df[['Country.1', 'Total KD5 Score of 20.1']].copy()

# Rename columns for consistency
df_2025.columns = ['Country', 'Total KD5 Score 2025', 'KD5 Score 2025']
df_2020.columns = ['Country', 'Total KD5 Score 2020']

# Merge the two dataframes on the country column
df_merged = pd.merge(df_2025, df_2020, on='Country')

# Calculate the difference between 2025 and 2020 Total KD5 Scores
df_merged['Score Difference'] = df_merged['Total KD5 Score 2025'] - df_merged['Total KD5 Score 2020']

# Plotting
plt.figure(figsize=(10, 8))

scatter = plt.scatter(
    df_merged['Total KD5 Score 2020'], 
    df_merged['Total KD5 Score 2025'], 
    s=df_merged['KD5 Score 2025'] * 20,  # Adjust size for better visualization
    c=df_merged['Score Difference'],  # Color based on the difference
    alpha=0.8,
    edgecolors="black",  # Set edge color to black for the dots
    cmap='Spectral'
)

# Adding labels and title
plt.xlabel('2020 Total KD5 Score of 20')
plt.ylabel('2025 Total KD5 Score of 20')
plt.title('Scatter Plot of KD5 Scores (2020 vs 2025)')

# Show 45-degree reference line
plt.plot([0, 22], [0, 22], ls="--", c=".3")

# Annotate points with country names and KD5 scores
texts = []
for i, country in enumerate(df_merged['Country']):
    x = df_merged['Total KD5 Score 2020'][i]
    y = df_merged['Total KD5 Score 2025'][i]
    texts.append(
        plt.text(
            x, y, f"{country} ({x:.1f}, {y:.1f})",
            fontsize=8,
            alpha=0.7
        xytext=(10, 10),  # Offset label slightly from the point
        textcoords='offset points',
        arrowprops=dict(arrowstyle="-", color="gray", alpha=0.7, lw=0.5)
    )

plt.colorbar(scatter, label='Score Difference (2025 - 2020)')
plt.grid(alpha=0.3)
plt.show()


# Load the Excel file and specify that the second row contains the column names
df = pd.read_excel('../../D1_jan2025/draft_january2025/Comparisons_2025_2020.xlsx', header=1)

# Extract relevant columns based on your data structure
df_2025 = df[['Country', 'Total KD5 Score of 20', 'KD5 Score']].copy()
df_2020 = df[['Country.1', 'Total KD5 Score of 20.1', 'KD5 Score.1']].copy()

# Rename columns for consistency
df_2025.columns = ['Country', 'Total KD5 Score 2025', 'KD5 Score 2025']
df_2020.columns = ['Country', 'Total KD5 Score 2020', 'KD5 Score ']

# Merge the two dataframes on the country column
df_merged = pd.merge(df_2025, df_2020, on='Country')
df_merged
```
